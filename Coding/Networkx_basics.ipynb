# Directed Graph

```python
import networkx as nx
```

```python
DG = nx.DiGraph()
```

```python
DG.add_nodes_from([1,2,3])
```

```python
DG.add_weighted_edges_from([(1,2,0.5),(3,1,0.75)])
```

```python
a = list(DG.successors(1))
print(a)
```

```python
a = list(DG.predecessors(1))
print(a)
```

```python
import matplotlib.pyplot as plt
```

```python
DG[1][2]['Name'] = 'red'
```

```python
nx.draw(DG, with_labels=True)
```

**pos = nx.spring_layout(DG)** ---> Gets 2D position of the nodes\
**edge_labels = nx.get_edge_attributes(DG, 'Name')** ---> Extracts labels from edges, use name used while defining labels like **weight, name, color, day, etc...**

```python
pos = nx.spring_layout(DG)

nx.draw(DG, pos, with_labels=True)

# 👇 Extract your custom attribute
edge_labels = nx.get_edge_attributes(DG, 'Name')

# 👇 Draw them
nx.draw_networkx_edge_labels(DG, pos, edge_labels=edge_labels)

plt.show()
```

**Convert DIRECTED to UNDIRECTED** --->  H = nx.Graph(DG)

```python
H = nx.Graph(DG)
```

```python
nx.draw(H)
```
